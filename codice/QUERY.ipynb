```python
import pandas as pd
from pymongo import MongoClient
```

```python
client= MongoClient('mongo', 27017, username='admin', password='DataMan2019!')
```

```python
db = list(client.dbintegrato.db_avgscore.find())
tweets = pd.DataFrame(db)
tweets.drop(['_id'], axis=1, inplace= True)
tweets.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

Effettuo una copia delle 2 colonne per la successiva lavorazione.

```python
tweets['IMDB_Rating_std'] = tweets['IMDB_Rating']
tweets['avg_score_std'] = tweets['avg_score']
tweets.head()
```

Standardizzo entrambe le colonne così da renderle confrontabili senza distorsioni dovute alla scala di valori.

```python
tweets[['IMDB_Rating_std','avg_score_std']] = StandardScaler().fit_transform(tweets[['IMDB_Rating_std', 'avg_score_std']])
tweets.head()
```

Verifico che la standardizzazione sia andata a buon fine:
dovrebbero risultare per entrambe le colonne media 0 e deviazione standard 1

```python
tweets[['IMDB_Rating_std','avg_score_std']].mean(axis = 0)
#media delle due colonne circa 0
```

```python
tweets[['IMDB_Rating_std','avg_score_std']].std(axis = 0)
#deviazione standard delle due colonne circa 1
```

Creo una colonna che metta in mostra la differenza tra i due rating per ogni film

```python
tweets['diff_std_rating'] = tweets['IMDB_Rating_std'] - tweets['avg_score_std']
tweets.head()
#in quelli positivi IMDB RATING è maggiore
#in quelli negativ IMDB RATING è minore
```

```python
tweets[['title','IMDB_Rating_std','avg_score_std','diff_std_rating']]
#le 3 colonne sono standardizzate, l'ultima rappresenta la differenza tra i due rating per ogni film
```

# Analisi dei generi

Il genere mediamente più apprezzato in entrambi i rating sembra essere Drama seguito da Adventure e Comedy.

Mostro la top 10 dei generi secondo la sentiment media dei tweet.

```python
IMDB_Genre = tweets.groupby('Genre')[['IMDB_Rating_std','avg_score_std']].mean()
IMDB_Genre.sort_values(by=['avg_score_std'], ascending=False).head(10)
```

Mostro la top 10 dei generi secondo IMDB rating.

```python
IMDB_Genre.sort_values(by=['IMDB_Rating_std'], ascending=False).head(10)
```

```python
IMDB_Genre['mean_rating'] = (IMDB_Genre['IMDB_Rating_std'] + IMDB_Genre['avg_score_std'])/2
```

Mostro la top 10 dei generi secondo la media di entrambi gli indicatori

```python
IMDB_Genre.sort_values(by=['mean_rating'], ascending=False).head(10)
```
